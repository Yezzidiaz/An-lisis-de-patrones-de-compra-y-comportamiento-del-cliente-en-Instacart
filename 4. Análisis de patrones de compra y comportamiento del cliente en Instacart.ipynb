### [A3] ¿Qué día de la semana compran víveres las personas?

```python
# Conteo de órdenes por día de la semana
pedidos_por_dia = orders['order_dow'].value_counts().sort_index()

# Etiquetas de los días
dias_semana = ['Domingo', 'Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado']

# Gráfico
plt.figure(figsize=(8,5))
pedidos_por_dia.plot(kind='bar')
plt.title('Cantidad de pedidos por día de la semana')
plt.xlabel('Día de la semana')
plt.ylabel('Cantidad de pedidos')
plt.xticks(ticks=range(7), labels=dias_semana, rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
reordenes = order_products.groupby('product_id')['reordered'].agg(['sum', 'count']).reset_index()
reordenes['reorder_rate'] = reordenes['sum'] / reordenes['count']
```

```python
top_reorder_rate = reordenes.sort_values(by='reorder_rate', ascending=False).head(20)
top_reorder_rate = top_reorder_rate.merge(products[['product_id', 'product_name']], on='product_id', how='left')
```

```python
print(top_reorder_rate)
```

Escribe aquí tus conclusiones.

Hay productos que casi siempre se vuelven a pedir después de la primera compra. Esto puede estar asociado con artículos indispensables, o que son difíciles de sustituir.

### [C4] Para cada cliente, ¿qué proporción de sus productos ya los había pedido?

```python
user_orders = orders[['order_id', 'user_id']].merge(order_products[['order_id', 'reordered']], on='order_id')
```

```python
user_reorder_rate = user_orders.groupby('user_id')['reordered'].agg(['sum', 'count'])
user_reorder_rate['user_reorder_rate'] = user_reorder_rate['sum'] / user_reorder_rate['count']
user_reorder_rate['user_reorder_rate'].hist(bins=50, figsize=(10,5))
plt.title('Distribución de tasa de repetición por usuario')
plt.xlabel('Tasa de repetición')
plt.ylabel('Número de usuarios')
plt.show()
```

Escribe aquí tus conclusiones.

La mayoría de los usuarios repiten productos con una tasa media-alta, lo que refuerza la idea de que la plataforma se usa frecuentemente para compras regulares. También hay usuarios que nunca repiten productos, lo cual podría indicar exploración, una sola compra o malas experiencias.
